### 문제3번
range() 함수와 유사한 frange() 함수를 다음 조건을 참고하여 만드시오.

1) frange()의 인자 구성은 다음과 같이 range와 동일하지만 각 인수들은 음수를 받지 않는다고 가정한다.
   - range(stop)
   - range(start, stop[, step])
   
2) frange() 함수의 인자에 대한 기본 시작(start) 값은 0.0이고, 기본 단계(step) 값은 0.1이다.

3) frange 사용 예
   - 3-1) frange(0.5)
       - [0.0, 0.1, 0.2, 0.3, 0.4]
   - 3-2) frange(1.0, 2.0)
       - [1.0, 1.1, 1.2, 1.3, 1.4, 1.5, 1.6, 1.7, 1.8, 1.9]
   - 3-3) frange(2.2, 4.0, 0.5)
       - [2.2, 2.7, 3.2, 3.7]

```python
def frange(stop, start = 0.0 ,step = 0.1):
    if stop < start :
        start, stop = stop, start
    l = []
    while start < stop:
        l.append(start)
        start = round((step+start),len(str(step))-2) # float 오차로인해
    return l
            

print frange(0.5)
print frange(1.0, 2.0)
print frange(2.2, 4.0, 0.5)
```

매개변수를 stop과 start, step으로 선언하고, start와 step에 디폴트 값을 준다. start가 stop보다 크면 그 값을 바꾸어 저장한다. 리스트 하나를 생성하고 while문을 통해 stop이 start보다 크면 리스트에 start 값을 추가하는데 단계만큼 증가시키면서 리스트에 값을 넣는다.

```python
def sum(*num):
    a = 0
    for x in num:
        a = a + x
    return a

print sum()
print sum(1, 2)
print sum(1, 2, 3, 4, 5)
print sum(1, 5, 7, 2, -10)
```

### 문제 5번
여러 단어로 이루어진 문자열을 입력받아 각 단어의 첫글자로 이루어진 단어를 대문자로 출력하는 myinitial() 함수를 다음 조건을 참고하여 작성하시오.

- 1) 다음에 제시되는 함수들을 모두 이용해야 한다.
   - split
   - map
   - join
- 2) myinitial() 함수 사용 예
   - myinitial("as soon as possible")
      - ASAP

```python
def myinitial(str):
    str = str.split(" ")
    str = map(lambda str:str[0], str)
    str = "".join(str)
    return str.upper()

print myinitial("as soon as possible")
```

myinitial()의 매개변수를 문자열로 입력받는다. 문자열을 공백기준으로 리스트를 만들어 저장한다. 문자열에 map()안에 존재하는 lambda함수를 사용하여 문자열의 원소의 첫번째 부분만 추출한것을 리스트로 만든다. 그리고 join을 사용해 추출한 문자열을 저장한다. 값을 출력할 때 대문자로 출력해야하므로 upper를 사용하면 된다.

### 문제 6번
음이 아닌 정수 n를 입력받으면 n! (factorial)을 계산하는 myfact() 함수를 재귀적 함수로 구현하시오.

- [옵션] 참을 수 있는 정도 만큼의 수행시간을 직접 기다려보면서 n을 늘려보도록 합시다.

```python
def fac(n):
    if n == 1:
        return n
    else:
        return n * fac(n-1)
    
print fac(10)
print fac(50)
print fac(200)
```

입력받은 정수의 n!를 재귀적으로 구하기위해 입력받은 수가 1이면 그대로 반환하고, 2 이상이면 자신의 함수를 호출하여 계산하도록 한다.

10 이하의 소수를 모두 더하면 2 + 3 + 5 + 7 = 17 이 됩니다.
이백만(2,000,000) 이하 소수의 합은 얼마입니까?

```python
def isprime(n, primes):
    for p in primes:
        if n % p == 0: 
            return 0
        if p * p > n: 
            return 1

sum = 0
primes = [2, ]
n = 3 
cnt = 1
while cnt < 2000001:
    if isprime(n,primes):
        primes.append(n)
        cnt = cnt + 1        
    n = n + 2  
for x in range(0, 2000001):
    sum = sum + primes[x]
print sum
```

### 문제 12번 (프로젝트 오일러 14번)
양의 정수 n에 대하여, 다음과 같은 계산 과정을 반복하기로 합니다.

    n → n / 2 (n이 짝수일 때)
    n → 3 n + 1 (n이 홀수일 때)

13에 대하여 위의 규칙을 적용해보면 아래처럼 10번의 과정을 통해 1이 됩니다.

    13 → 40 → 20 → 10 → 5 → 16 → 8 → 4 → 2 → 1
아직 증명은 되지 않았지만, 이런 과정을 거치면 어떤 수로 시작해도 마지막에는 1로 끝나리라 생각됩니다. 
(역주: 이것은 콜라츠 추측 Collatz Conjecture이라고 하며, 이런 수들을 우박수 hailstone sequence라 부르기도 합니다)

그러면, 백만(1,000,000) 이하의 수로 시작했을 때 1까지 도달하는데 가장 긴 과정을 거치는 숫자는 얼마입니까?

참고: 계산 과정 도중에는 숫자가 백만을 넘어가도 괜찮습니다.

```python
def collatz(n):
    count = 1
    while n > 1:
        count += 1
        if n % 2 == 0:
            n = n/2
        else:
            n = 3*n + 1
    return count
 
max = [0,0]
for i in range(1000000):
    c = collatz(i)
    if c > max[0]:
        max[0] = c
        max[1] = i
print max[1]
```

                - similarity = {}

                - similarity[url2] = x2
                - ...
      - 유사도가 높은 웹 사이트가 상위에 출력되어야 함 (즉, 유사도 기준 내림 차순)

   - 7) [주의] 필수사항
     - 위에서 만든 검색엔진 코딩은 매우 간단한 것이라 부족한 점이 많이 존재한다.
     - 본인이 생각하기에 상업적인 완성도 높은 검색 로봇/엔진이 되려면 어떤 기능들이 추가적으로 구현되어야 할지 최소 3가지 이상 제시하시오.

```python


import urllib2
import string
import pickle



def tag_delete(string):
    for i in range(0, string.count('<!--')):
        if string.find("<!--") != -1:
            string = string.replace(string[string.find("<!--"):string.find("-->")+3:]," ")
    for i in range(0,string.count("<")):
        if string.find("<") != -1:
            string = string.replace(string[string.find("<"):string.find(">")+1:]," ")
    
    return string

#자바스크립트 제거
def javascript_delete(string):
    for i in range(0, string.count('<script')):
        if string.find("<script") != -1:
            string = string.replace(string[string.find("<script"):string.find("</script>")+9:]," ")
    return string

#css 제거
def stylesheet_delete(string):
    for i in range(0, string.count('<style')):
        if string.find("<style") != -1:
            string = string.replace(string[string.find("<style"):string.find("</style>")+8:]," ")
    return string

#구두문자 제거 함수
def punc_delete(source):
    punctuation = string.punctuation
    for i in range (len(punctuation)):
        source= source.replace(punctuation[i]," ")
        
    return source

#불용어 제거 함수
def stopword_delete(words):
    stopword = [ 'A','a', 'and', 'are', 'as', 'at', 'be', 'but','But', 'by', 'for', 'if','If', 'in', 'into', 'is', 'it','It', 'no','No', 'not','Not', 'of', 'on', 'or', 's', 'such', 't', 'that','That', 'the','The', 'their','Their', 'then','Then', 'there','There', 'these','These', 'they','They', 'this','This', 'to', 'was', 'will', 'with']
    for i in range (len(stopword)):
        for j in range(len(words)):
            if words[j] == stopword[i]:
                words[j] = " "


#사전형태로 출현빈도 저장 함수
def word_count(words):
    words_dict = {}
    for i in words:
        if words_dict.has_key(i) is False:
            words_dict[i]=1
        else:
            words_dict[i]+=1
    return words_dict




f = open('www.cnn.com.html','w')
f.write(urllib2.urlopen("http://www.cnn.com").read()) # 문자열을 파일에 기록
f.close()

f = file('www.cnn.com.html') # f = open('t.txt', 'r')과 동일
source = f.read()
f.close()

source = stylesheet_delete(source)
source = javascript_delete(source)
source = tag_delete(source)
source = punc_delete(source)
words = source.split()  
stopword_delete(words)
source = " ".join(words)
words = source.split()
words = word_count(words)

f = open('www.cnn.com.words.frequency.txt', 'w')
pickle.dump(words, f)
f.close()




f = open('www.github.com.html','w')
f.write(urllib2.urlopen("http://www.github.com").read()) # 문자열을 파일에 기록
f.close()

f = file('www.github.com.html') # f = open('t.txt', 'r')과 동일
source = f.read()
f.close()

source = stylesheet_delete(source)
source = javascript_delete(source)
source = tag_delete(source)
source = punc_delete(source)
words = source.split()  
stopword_delete(words)
source = " ".join(words)
words = source.split()
words = word_count(words)

f = open('www.github.com.words.frequency.txt', 'w')
pickle.dump(words, f)
f.close()





f = open('www.time.com.html','w')
f.write(urllib2.urlopen("http://www.time.com").read()) # 문자열을 파일에 기록
f.close()

f = file('www.time.com.html') # f = open('t.txt', 'r')과 동일
source = f.read()
f.close()

source = stylesheet_delete(source)
source = javascript_delete(source)
source = tag_delete(source)
source = punc_delete(source)
words = source.split()  
stopword_delete(words)
source = " ".join(words)
words = source.split()
words = word_count(words)

f = open('www.time.com.words.frequency.txt', 'w')
pickle.dump(words, f)
f.close()



f = open('www.nasa.gov.html','w')
f.write(urllib2.urlopen("http://www.nasa.gov").read()) # 문자열을 파일에 기록
f.close()

f = file('www.nasa.gov.html') # f = open('t.txt', 'r')과 동일
source = f.read()
f.close()

source = stylesheet_delete(source)
source = javascript_delete(source)
source = tag_delete(source)
source = punc_delete(source)
words = source.split()  
stopword_delete(words)
source = " ".join(words)
words = source.split()
words = word_count(words)

f = open('www.nasa.gov.words.frequency.txt', 'w')
pickle.dump(words, f)
f.close()



f = open('www.nbc.com.html','w')
f.write(urllib2.urlopen("http://www.nbc.com").read()) # 문자열을 파일에 기록
f.close()

f = file('www.nbc.com.html') # f = open('t.txt', 'r')과 동일
source = f.read()
f.close()

source = stylesheet_delete(source)
source = javascript_delete(source)
source = tag_delete(source)
source = punc_delete(source)
words = source.split()  
stopword_delete(words)
source = " ".join(words)
words = source.split()
words = word_count(words)

f = open('www.nbc.com.words.frequency.txt', 'w')
pickle.dump(words, f)
f.close()




f = open('www.nbc.com.words.frequency.txt')
bbc = pickle.load(f)
f.close()

f = open('www.nasa.gov.words.frequency.txt')
sciencemag = pickle.load(f) # 인스턴스 가져오기
f.close()

f = open('www.time.com.words.frequency.txt')
time = pickle.load(f) # 인스턴스 가져오기
f.close()

f = open('www.github.com.words.frequency.txt')
about = pickle.load(f) # 인스턴스 가져오기
f.close()

f = open('www.cnn.com.words.frequency.txt')
cnn = pickle.load(f) # 인스턴스 가져오기
f.close()



combine ={}
for i in bbc:
    combine[i] = bbc[i]
    
for i in sciencemag:
    if combine.has_key(i):
        combine[i] += sciencemag[i]
    else:
        combine[i] = sciencemag[i]

for i in time:
    if combine.has_key(i):
        combine[i] += time[i]
    else:
        combine[i] = time[i]

for i in about:
    if combine.has_key(i):
        combine[i] += about[i]
    else:
        combine[i] = about[i]
        
for i in cnn:
    if combine.has_key(i):
        combine[i] += cnn[i]
    else:
        combine[i] = cnn[i]
        
        
g = combine.items()
g.sort(key=lambda x:x[1],reverse=True)
for i in range(3):
    print g[i]

```
