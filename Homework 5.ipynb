    - count: Counter 인스턴스가 지니는 초기 정수 값
    - step: Counter 인스턴스의 count가 증가되는 증분 (defalt 값: 1)

- 요구사항 2. 다음과 같이 Counter의 인스턴스를 출력을 해주는 __str__() 메소드를 Counter 클래스 내에 구현하시오.

- 요구사항 3. 다음과 같이 step에 주어진 증분만큼 count를 증가시키는 incr() 메소드를 Counter 클래스 내에 구현하시오.

        >>> print c 
        [Count (step: 1)] 11 
        >>> print d 
        [Count (step: 2)] 12 
        
- 요구사항 4. Counter 클래스 내에 관련 메소드를 추가하여 인스턴스 객체를 직접 호출(call)할 수 있도록 하시오. 인스턴스 객체를 직접 호출했을 때에 내부적으로 incr() 메소드를 호출하는 방법으로 구현하시오.

        >>> c() 
        >>> d() 
        >>> print c 
        [Count (step: 1)] 12 
        >>> print d 
        [Count (step: 2)] 14 
        
- 요구사항 5. 다음과 같은 두 개의 산술 연산 (+, -)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.

        >>> c = c + 5 
        >>> d = d - 5 
        >>> print c 
        [Count (step: 1)] 17 
        >>> print d 
        [Count (step: 2)] 9 
        
- 요구사항 6. 다음과 같은 관계연산 (>, <, ==)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.

        >>> print c > 10 
        True 
        >>> print d > 10 
        False 
        >>> print c < 10 
        False 
        >>> print d < 10 
        True 
        >>> print c == 17 
        True 
        >>> print d != 9 
        False 

```python
class Counter():
    def __init__(self, count, step = 1):
        self.count = count
        self.step = step

    def __str__(self):
        return '[Count (step : ' + str(self.step) + ')] ' + str(self.count)
    
    def incr(self):
        self.count += self.step
        return self
    
    def __call__(self):
        return self.incr()
    
    def __add__(self, other):
        self.count += other
        return self
    
    def __sub__(self, other):
        self.count -= other
        return self
    
    def __lt__(self, y): # <
        return self.count < y
    def __gt__(self, y): #>
        return self.count > y
    def __eq__(self, y): #==
        return self.count == y
    def __ne__(self, y): #==
        return self.count != y    
    
c = Counter(10) 
d = Counter(10, 2)

print c
print d
```

#### 문제4. 다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 __init__(), __str()__(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.

- MySet은 집합(Set) 자료형을 정의하려는 의도하에 만들어진 클래스이다.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print s
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print t
```

init은 클래스의 생성자이다. 생성자는 list를 매개변수를 통하여 리스트의 요소들을 클래스 내부에 추가한다. 그리고 eliminateduplicate 함수를 호출하도록 되어있다. str은 출력문을 위해 존재하는 함수이다. str 함수는 클래스 내부의 리스트를 문자열로 출력하도록 설계되어있다. 마지막으로 dliminate_duplicate 함수는 리스트를 매개변수로 하여 리스트 내에 중복이 존재하는 원소들은 제거한다.

#### 문제5. 4번 문제에 정의된 MySet 클래스에 메소드를 추가하여 다음 각 요구사항 모두를 만족시키는 코딩을 제시하시오

- 정답을 각 요구사항별로 입력할 필요 없이 요구사항 3개 전체에 대해 1개의 MySet 클래스 정의 코드를 제시하고 주석등으로 해당 코드를 설명한다.
- 요구사항 1. | 연산으로 두 집합의 합집합을 반환한다.

    - u = s | t 
    - print u 
    - MySet: {1, 2, 3, 4, 5, 6, 7, 8, 9} 
- 요구사항 2. & 연산으로 두 집합의 교집합을 반환한다.

    - u = s & t 
    - print u 
    - MySet: {2, 3} 

- 요구사항 3. - 연산으로 두 집합의 차집합을 반환한다.

    - s = MySet([1, 2, 3]) 
    - t = MySet([3, 4, 5]) 
    - u = s - t 
    - print u 
    - MySet: {1} 

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
            
    def __or__(self, l):
        list2 = []
        for e in l:
            list2.append(e)
        for e in self:
            if e not in list2:
                list2.append(e)
        return MySet(l2)
    
    def __and__(self, l):
        list2 = []
        for e in l:
            if e in self:
                list2.append(e)
        return MySet(list2)
        
    def __sub__(self, l):
        list2 = []
        for e in self:
            list2.append(e)
        for e in list2:
            if e in l:
                list2.remove(e)
        return MySet(list2)
        
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print s
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print t
    
    u = s|t
    print u
    
    u = s&t
    print u
    
    s = MySet([1, 2, 3])
    t = MySet([3, 4, 5])
    u = s - t
    print u
```

위에서 제시된 소스코드에 각 집합과 관련된 함수들을 조건에 맞게 정의하고 코딩하였다.
