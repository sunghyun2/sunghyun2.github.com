     s = ' first star : second star : third star '

- 2) 리스트 내포(list comprehension)를 이용한 코드를 작성하시오.

```python
s = ' first star : second star : third star '
lens = s.count(':') + 1
list = ['' for a in range(0, lens)]

n = 0
for x in range(0, len(s)):
    if s[x] == ':':
        n = n +1
        continue
    list[n] = list[n] + s[x]
for x in range(0, len(list)):
    list[x] = list[x].strip()
print list
```

#### 문제5. 다음과 같이 0보다 큰 정수 리스트 변수 list를 인자로 받는 함수 addall(list)와 addallodd(list)를 for ~ in 리터널과 리스트 내포 방식으로 각각 제시하시오.
1. addall(list) 함수 (리스트 내의 모든 정수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)
       >>> addall([1])
       1
       >>> addall([1, 2, 3, 4, 5, 6, 7, 8, 9])
       45
2. addallodd(list) 함수 (리스트내의 모든 홀수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)
       >>> addallodd([1])
       1
       >>> addallodd([1, 2, 3, 4, 5, 6, 7, 8, 9])
       25
 
 - [참고 1]: 리스트 내포 방식으로 addall() 및 addallodd()함수를 만들 때엔 리스트 내포에 의해 생성된 리스트 결과에 대해 내장 함수를 사용하는 방식을 고려해보시오 (주의: 리스트 내포 방식의 addall() 및 addallodd() 함수의 몸체는 단 1줄로 작성되어야 한다.)
 - [참고 2]: sum() 등의 내장함수 사용 가능함

```python
def addall(list1):
    sum = 0
    for x in range (0, len(list1)):
        sum = sum + list1[x]
    return sum

def addallodd(list2):
    sum = 0
    for x in range(0, len(list2)):
        if(list2[x] % 2 == 1):
            sum = sum + list2[x]
    return sum

list1 = [1]
list2 = [1,2,3,4,5,6,7,8,9]

print addall(list1)
print addall(list2)
print addallodd(list1)
print addallodd(list2)
```

- addall 함수는 for ~ in을 사용해 리스트의 첫 원소부터 마지막 원소까지 더하여 sum에 넣어주었다. 
- addallodd 함수는 리스트 원소 중 홀수만 더 하는 것으로 if x%2가 1이면 홀수이므로 해당되는 원소만 더해진다. 

```python
def addall(list):
    return sum([x for x in list])

def addallodd(list):
    return sum([x for x in list if x % 2])

list1 = [1]
list2 = [1, 2, 3, 4, 5, 6, 7, 8, 9]

print addall(list1)
print addall(list2)

print addallodd(list1)
print addallodd(list2)
```

- addall 함수를 내포를 사용하여 한 줄로 표현하기 위해 return 문에 한 번에 넣었다.
- addallodd 함수는 리스트의 원소들 중 홀수만 표현해야 하므로 if문을 한줄에 표현한다. if x%2가 참이면 소스가 동작한다.
- 두 함수 모두 sum()이라는 내장함수를 사용하여 한줄로 표현할 수 있다.
